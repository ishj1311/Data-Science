```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("iris.csv")
dataset.head() 
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,9))
sns.histplot(dataset['SepalLengthCm'], ax = axes[0,0])
sns.histplot(dataset['SepalWidthCm'], ax = axes[0,1])
sns.histplot(dataset['PetalLengthCm'], ax = axes[1,0])
sns.histplot(dataset['PetalWidthCm'], ax = axes[1,1])
```

```python
fig,axes = plt.subplots(2,2,figsize=(16,9))
sns.boxplot(y='SepalLengthCm', x='Species', data=dataset, ax=axes[0,0])
sns.boxplot(y='SepalWidthCm', x='Species', data=dataset, ax=axes[0,1])
sns.boxplot(y='PetalLengthCm', x='Species', data=dataset, ax=axes[1,0])
sns.boxplot(y='PetalWidthCm', x='Species', data=dataset, ax=axes[1,1])
```
