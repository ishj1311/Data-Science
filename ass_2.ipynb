```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv('academic.csv')
df
```

```python
df.isnull().sum()
```

```python
df['Name'] = df['Name'].fillna('Missing')

df['Attendance'] = df['Attendance'].fillna(df['Attendance'].mean())
df['CS'] = df['CS'].fillna(0)
df
```

```python
z= np.abs(stats.zscore(df['Attendance']))
z
```

```python
outlier = np.where(z>2)
outlier
```

```python
df = df.drop(outlier[0])
df
```

```python
df['Attendance'] = np.log(df['Attendance'])
df
```
