```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
df = pd.read_csv("Iris.csv")
df
```

```python
grouped_data = df.groupby("Species")
```

```python
summary_data = grouped_data.agg({'PetalWidthCm': ['mean', 'median', 'min', 'max', 'std'],
                                 'PetalLengthCm': ['mean', 'median', 'min', 'max', 'std'],
                                'SepalWidthCm' : ['mean', 'median', 'min', 'max', 'std'],
                                'SepalLengthCm': ['mean', 'median', 'min', 'max', 'std']})
summary_data
```

Percentile:-

```python
summary_data = grouped_data.agg({
    'PetalWidthCm': [
        ('25%', lambda x: x.quantile(0.25)),
        ('50%', lambda x: x.quantile(0.5)),
        ('75%', lambda x: x.quantile(0.75))
    ],
    'PetalLengthCm': [
        ('25%', lambda x: x.quantile(0.25)),
        ('50%', lambda x: x.quantile(0.5)),
        ('75%', lambda x: x.quantile(0.75))
    ],
    'SepalWidthCm': [
        ('25%', lambda x: x.quantile(0.25)),
        ('50%', lambda x: x.quantile(0.5)),
        ('75%', lambda x: x.quantile(0.75))
    ],'SepalLengthCm': [
        ('25%', lambda x: x.quantile(0.25)),
        ('50%', lambda x: x.quantile(0.5)),
        ('75%', lambda x: x.quantile(0.75))
    ]
})
summary_data
```
